```python
from flask import Flask, request, render_template_string
import webbrowser

app = Flask(__name__)

# 🧠 AI Logic
def priority(s):
    p = 1
    if s["difficulty"] == "hard":
        p += 2
    elif s["difficulty"] == "medium":
        p += 1

    if s["weak"]:
        p += 2

    if s["days"] <= 5:
        p += 2

    return p


def plan(subjects, total):
    total_p = sum(priority(s) for s in subjects)
    for s in subjects:
        s["time"] = round((priority(s) / total_p) * total, 2)
    return subjects


# 🌐 HTML UI
HTML = """
<!DOCTYPE html>
<html>
<head>
<title>AI Study Planner</title>
<style>
body { font-family: Arial; background:#1e1e2f; color:white; text-align:center; }
input, select { padding:8px; margin:5px; }
button { padding:10px 20px; background:green; color:white; border:none; }
</style>
</head>
<body>

<h1>AI Study Planner</h1>

<form method="post">
<input name="name" placeholder="Subject"><br>

<select name="difficulty">
<option>easy</option>
<option>medium</option>
<option>hard</option>
</select><br>

Weak? <input type="checkbox" name="weak"><br>

<input name="days" placeholder="Days left"><br>
<input name="hours" placeholder="Total hours"><br>

<button type="submit">Generate</button>
</form>

{% if result %}
<h2>Result:</h2>
{% for r in result %}
<p>{{r}}</p>
{% endfor %}
{% endif %}

</body>
</html>
"""

@app.route("/", methods=["GET", "POST"])
def home():
    result = []

    if request.method == "POST":
        s = {
            "name": request.form["name"],
            "difficulty": request.form["difficulty"],
            "weak": "weak" in request.form,
            "days": int(request.form["days"])
        }

        total = float(request.form["hours"])
        data = plan([s], total)

        for i in data:
            result.append(f"{i['name']} → {i['time']} hrs")

    return render_template_string(HTML, result=result)


if __name__ == "__main__":
    webbrowser.open("http://127.0.0.1:5000/")
    app.run()
```
